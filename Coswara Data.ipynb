```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('combined_data.csv')
```

```python
freq_by_gender = df['g'].value_counts()
freq_by_gender.plot.pie(autopct="%.1f%%")
plt.title('Patients by gender')
```

```python
ages = []
ages_male = []
ages_female = []
for i in range(0, 17):
    start_age = i*5
    end_age = i*5 + 4
    by_age = df[(df.a >= start_age) & (df.a <= end_age)]
    age_counts = by_age['g'].value_counts()
    ages.append(str(start_age) + '-' + str(end_age))
    if ('male' in age_counts):
        ages_male.append(-age_counts.male)
    else:
        ages_male.append(0)
    if ('female' in age_counts):
        ages_female.append(age_counts.female)
    else:
        ages_female.append(0)
by_age_df = pd.DataFrame({
    'Age': ages,
    'Male': ages_male,
    'Female': ages_female
})
bar_plot = sns.barplot(x = 'Male', y = 'Age', data = by_age_df)
bar_plot = sns.barplot(x = 'Female', y = 'Age', data = by_age_df)
bar_plot.set(xlabel="Patients", ylabel="Age Group", title = "Population Pyramid")
```
